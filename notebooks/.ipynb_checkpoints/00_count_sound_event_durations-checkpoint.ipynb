```python
import csv
from os import listdir
from os.path import isfile, join
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib.cm import get_cmap
from mpl_toolkits.mplot3d import axes3d, Axes3D
font = {'family' : 'sans-serif',
        'size'   : 12}

rc('font', **font)
```

```python
folds = ['train', 'test', 'validate']
folder_data = '/data_ssd/users/pzinemanas/MAVD-traffic/annotations'
durations = {}

for fold in folds:
    durations[fold] = {}
    folder = join(folder_data,fold)
    onlyfiles = [f for f in listdir(folder) if (isfile(join(folder, f)))]

    for file in onlyfiles:
        file_label = join(folder,file)
        #print(file_label)
        with open(file_label, 'r') as csvfile:
            spamreader = csv.reader(csvfile, delimiter='\t')
            for row in spamreader:
                time = float(row[1])-float(row[0])
                event_label = row[2]
                if event_label in durations[fold]:
                    durations[fold][event_label] += time
                else:
                    durations[fold][event_label] = time  
    print('\n',fold)
    for keys,values in durations[fold].items():
        print(keys,values, 'sec')    
```

```python
list1 = ['car','bus','motorcycle','truck']
list2 = ['engine_idling','engine_accelerating','brakes','wheel_rolling','compressor']
list1_plot = ['car','bus','moto.','truck']
list2_plot = ['engine\nidling','engine\nacc.','brakes','wheel','comp.']

list3 = ['car/engine_idling','car/engine_accelerating', 'car/brakes','car/wheel_rolling',
              'bus/engine_idling', 'bus/engine_accelerating', 'bus/brakes','bus/compressor','bus/wheel_rolling',
              'motorcycle/engine_idling','motorcycle/engine_accelerating', 'motorcycle/brakes']

list3_plot = ['Car\nengine\nidling','Car\nengine\nacc.', 'Car\nbrakes','Car\nwheel\nrolling',
              'Bus\nengine\nidling', 'Bus\nengine\nacc.', 'Bus\nbrakes','Bus\ncomp.','Bus\nwheel\nrolling',
              'Moto\nengine\nidling','Moto\nengine\nacc.', 'Moto\nbrakes']
M_train = np.zeros((4,5))
M_test = np.zeros((4,5))
M_val = np.zeros((4,5))

M_train1 = np.zeros(4)
M_train2 = np.zeros(5)
M_train3 = np.zeros(len(list3))
M_test1 = np.zeros(4)
M_test2 = np.zeros(5)
M_test3 = np.zeros(len(list3))
M_val1 = np.zeros(4)
M_val2 = np.zeros(5)
M_val3 = np.zeros(len(list3))
for i in range(M_train.shape[0]):
    for j in range(M_train.shape[1]):
        clase = list1[i]+'/'+list2[j]
        if clase in durations['train']:
            M_train[i,j] = durations['train'][clase]
            M_train1[i] += durations['train'][clase]
            M_train2[j] += durations['train'][clase]
            if clase in list3:
                ix = list3.index(clase)
                M_train3[ix] = durations['train'][clase]
        if clase in durations['test']:
            M_test[i,j] = durations['test'][clase]
            M_test1[i] += durations['test'][clase]
            M_test2[j] += durations['test'][clase]
            if clase in list3:
                ix = list3.index(clase)
                M_test3[ix] = durations['test'][clase]
        if clase in durations['validate']:
            M_val[i,j] = durations['validate'][clase]
            M_val1[i] += durations['validate'][clase]
            M_val2[j] += durations['validate'][clase]
            if clase in list3:
                ix = list3.index(clase)
                M_val3[ix] = durations['validate'][clase]
            
cmap = get_cmap('viridis')


plt.figure(figsize=(18,4))

ax1 = plt.subplot2grid((1, 4), (0, 0))
ax2 = plt.subplot2grid((1, 4), (0, 1))
ax3 = plt.subplot2grid((1, 4), (0, 2), colspan=2)

ax1.set_title('Vehicles level')
ax1.bar(np.arange(len(list1))+0.5, M_train1, color=cmap(0.2), tick_label=list1_plot, label='train')#,log=True)
ax1.bar(np.arange(len(list1))+0.5, M_val1, bottom=M_train1, color=cmap(0.5), tick_label=list1_plot, label='validate')
ax1.bar(np.arange(len(list1))+0.5, M_test1, bottom=M_train1+M_val1, color=cmap(0.7), tick_label=list1_plot,label='test')
ax1.set_ylabel('time(s)')
#ax1.set_xticks(np.arange(len(list1))+0.9)
ax1.grid()

ax2.set_title('Components level')
ax2.bar(np.arange(len(list2))+0.5, M_train2, color=cmap(0.2), tick_label=list2_plot, label='train')#,log=True)
ax2.bar(np.arange(len(list2))+0.5, M_val2, bottom=M_train2, color=cmap(0.5), tick_label=list2_plot, label='validate')
ax2.bar(np.arange(len(list2))+0.5, M_test2, bottom=M_train2+M_val2, color=cmap(0.7), tick_label=list2_plot,label='test')
ax2.set_ylabel('time(s)')
#ax2.set_xticks(np.arange(len(list2))+0.9)
ax2.grid()

ax3.set_title('Subordinate level')
ax3.bar(np.arange(len(list3))+0.5, M_train3, color=cmap(0.2), tick_label=list3_plot, label='train')#,log=True)
ax3.bar(np.arange(len(list3))+0.5, M_val3, bottom=M_train3, color=cmap(0.5), tick_label=list3_plot, label='validate')#,log=True)
ax3.bar(np.arange(len(list3))+0.5, M_test3, bottom=M_train3+M_val3, color=cmap(0.7), tick_label=list3_plot,label='test')#,log=True)
ax3.set_ylabel('time(s)')
#ax3.set_xticks(np.arange(len(list3))+0.9)
ax3.grid()

plt.legend(loc='upper right')
plt.tight_layout()
#plt.savefig("MAVD_count_total.eps")
plt.show()
```
